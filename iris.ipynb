```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.30)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
#Model creation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=7)
nb=GaussianNB()
sv=SVC()
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
```

```python
z=[knn,nb,sv]
for i in z:
  print("Model is:",i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("******************************************")
  print("confusion matrix is \n",confusion_matrix(y_test,y_pred))
  lab=['Iris-versicolor','Iris-virginica','Iris-setosa']#enter the label data in order of y_train
  display=ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred),display_labels=lab)
  print(display.plot())
  print("accuracy is \n",accuracy_score(y_test,y_pred))
  print("classification report \n", classification_report(y_test,y_pred))
  print("*"*100)
```
